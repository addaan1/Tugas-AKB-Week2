# Analisis Framing Bisnis Olist Dataset
**Framing 1: Masalah utama adalah keterlambatan pengiriman (delivery delay).**

Notebook ini berisi langkah-langkah untuk memproses dataset E-Commerce Olist untuk menetapkan kerangka keputusan bisnis berdasarkan instruksi tugas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_theme(style="whitegrid")
```

## 1. Memuat Dataset

```python
try:
    orders = pd.read_csv('olist_orders_dataset.csv')
    items = pd.read_csv('olist_order_items_dataset.csv')
    payments = pd.read_csv('olist_order_payments_dataset.csv')
    reviews = pd.read_csv('olist_order_reviews_dataset.csv')
    print("Semua dataset berhasil dimuat.")
except FileNotFoundError as e:
    print("Dataset tidak lengkap atau tidak ditemukan.\nError:", e)
```

## 2. Pembangunan Dataset (Level-Order)

```python
# Agregasi items, payments, dan reviews per order_id
order_items = items.groupby('order_id').agg({'price': 'sum', 'freight_value': 'sum'}).reset_index()
order_payments = payments.groupby('order_id').agg({'payment_value': 'sum'}).reset_index()
order_reviews = reviews.groupby('order_id').agg({'review_score': 'mean'}).reset_index()

# Merge
df = orders.merge(order_items, on='order_id', how='left')
df = df.merge(order_payments, on='order_id', how='left')
df = df.merge(order_reviews, on='order_id', how='left')

# Konversi Datetime
date_cols = ['order_purchase_timestamp', 'order_delivered_customer_date', 'order_estimated_delivery_date']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# Drop pesanan yang belum/tidak terkirim
df = df.dropna(subset=['order_delivered_customer_date'])
print(f"Total data order unik yang berhasil diproses: {df.shape[0]}")
```

## 3. Pembuatan Variabel Turunan Sesuai Persyaratan

```python
# Variabel 1: Delivery Duration (Durasi Pengiriman aktual)
df['delivery_duration'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.days

# Variabel 2: Timeliness Status (Ketepatan Waktu)
df['timeliness_status'] = np.where(df['order_delivered_customer_date'] > df['order_estimated_delivery_date'], 'Barang Terlambat', 'Tepat Waktu')

# Kategori keterlambatan lebih rinci untuk analitik (Delivery Duration Status)
def categorize_duration(row):
    if row['order_delivered_customer_date'] <= row['order_estimated_delivery_date']:
        return 'Lebih Cepat/Tepat Estimasi'
    diff = (row['order_delivered_customer_date'] - row['order_estimated_delivery_date']).days
    if diff <= 7:
        return 'Terlambat Ringan (1-7 Hari)'
    else:
        return 'Terlambat Parah (>7 Hari)'

df['delivery_duration_kategori'] = df.apply(categorize_duration, axis=1)

display(df[['order_id', 'delivery_duration', 'timeliness_status', 'delivery_duration_kategori']].head())
```

## 4. Requirement 1: Grafik Baseline 
*(Tren tingkat ketepatan waktu / rasio pesanan yang terlambat per bulan yang menunjukkan korelasi negatif nyata dengan ulasan).* 

```python
# Agregasi Trend Waktu (Bulanan)
df['purchase_month'] = df['order_purchase_timestamp'].dt.to_period('M').astype(str)

# Filter outlier untuk bulan awal dengan volume kecil (misalnya sebelum 2017-01)
df_filtered = df[df['order_purchase_timestamp'] >= '2017-01-01'].copy()

# Hitung Late Rate (%) dan Average Review Score per bulan
monthly_stats = df_filtered.groupby('purchase_month').agg(
    late_rate=('timeliness_status', lambda x: (x == 'Barang Terlambat').mean() * 100),
    avg_review=('review_score', 'mean')
).reset_index()

fig, ax1 = plt.subplots(figsize=(14, 7))

# Bar chart untuk Late Rate 
ax1.bar(monthly_stats['purchase_month'], monthly_stats['late_rate'], color='salmon', alpha=0.7, label='Terlambat (%)')
ax1.set_xlabel('Bulan Pemesanan', fontsize=12, fontweight='bold')
ax1.set_ylabel('Persentase Pesanan Terlambat (%)', color='salmon', fontsize=12, fontweight='bold')
ax1.tick_params(axis='y', labelcolor='salmon')
plt.xticks(rotation=45)

# Line chart untuk Rata-rata Skor Ulasan
ax2 = ax1.twinx()  
ax2.plot(monthly_stats['purchase_month'], monthly_stats['avg_review'], color='darkblue', marker='o', linewidth=2.5, label='Rata-Rata Review')
ax2.set_ylabel('Skor Ulasan (Review Score 1-5)', color='darkblue', fontsize=12, fontweight='bold')
ax2.tick_params(axis='y', labelcolor='darkblue')

plt.title('Baseline: Tren Korelasi Negatif antara Tingkat Keterlambatan Pengiriman dan Kepuasan Pelanggan', fontsize=14, fontweight='bold')

fig.legend(loc='upper right', bbox_to_anchor=(0.90, 0.85))
plt.show()
```

## 5. Requirement 2: Dua Tabel Komparatif
Membuktikan secara kuantitatif betapa rusaknya efek pengiriman terlambat terhadap performa bisnis.

```python
print("=======================================================================================")
print("Tabel 2.1: Distribusi Penilaian Pelanggan terhadap Status Ketepatan Waktu Pengiriman")
print("=======================================================================================")

# Kolom Flag ulasan 1 bintang
df['is_review_1_bintang'] = (df['review_score'] <= 1.0).astype(int)

tabel_1 = df.groupby('timeliness_status').agg(
    Jumlah_Order=('order_id', 'count'),
    Rata2_Review=('review_score', 'mean'),
    Persentase_Review_1_Bintang=('is_review_1_bintang', lambda x: x.mean() * 100)
).reset_index()

# Format Tabel 1
tabel_1['Persentase_Total_Order (%)'] = (tabel_1['Jumlah_Order'] / tabel_1['Jumlah_Order'].sum() * 100).round(2)
tabel_1['Rata2_Review'] = tabel_1['Rata2_Review'].round(2)
tabel_1['Persentase_Review_1_Bintang (%)'] = tabel_1['Persentase_Review_1_Bintang'].round(2)

# Menyusun ulang urutan kolom ke depan
tabel_1 = tabel_1[['timeliness_status', 'Persentase_Total_Order (%)', 'Rata2_Review', 'Persentase_Review_1_Bintang (%)']]
display(tabel_1)

print("\n\n=====================================================================================================")
print("Tabel 2.2: Rata-Rata Penilaian terhadap Tingkat Keparahan Keterlambatan (Delivery Duration Difference)")
print("=====================================================================================================")

tabel_2 = df.groupby('delivery_duration_kategori').agg(
    Jumlah_Order=('order_id', 'count'),
    Rata_Rata_Waktu_Tiba_Aktual_Hari=('delivery_duration', 'mean'),
    Rata2_Review=('review_score', 'mean'),
    Persentase_Review_1_Bintang=('is_review_1_bintang', lambda x: x.mean() * 100)
).reindex(['Lebih Cepat/Tepat Estimasi', 'Terlambat Ringan (1-7 Hari)', 'Terlambat Parah (>7 Hari)']).reset_index()

tabel_2['Rata_Rata_Waktu_Tiba_Aktual_Hari'] = tabel_2['Rata_Rata_Waktu_Tiba_Aktual_Hari'].round(1)
tabel_2['Rata2_Review'] = tabel_2['Rata2_Review'].round(2)
tabel_2['Persentase_Review_1_Bintang (%)'] = tabel_2['Persentase_Review_1_Bintang'].round(2)
display(tabel_2)
```
